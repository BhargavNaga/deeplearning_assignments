```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the dataset
dataset = pd.read_csv("lin_reg.csv")
x = dataset.iloc[:, 0].values
y = dataset.iloc[:, -1].values
print(x.shape, y.shape)
print(x[0])
print(y[0])
```

```python
# Function to compute Mean Squared Error
def compute_cost(y, y_pred):
    return np.mean((y_pred - y) ** 2)

# Function to perform gradient descent
def gradient_descent(x, y, learning_rate=0.0001, iterations=100):
    num_samples = len(y)
    m = 0  # Slope
    c = 0  # Y-intercept
    costs = []

    for i in range(iterations):
        y_pred = m * x + c


        # Compute gradients
        dm = (2/num_samples) * np.dot(x.T, (y_pred - y))
        dc = (2/num_samples) * np.sum(y_pred - y)
        print(f"dm = {dm}, dc = {dc}") if i == 0 else None
        # Update parameters
        m -= learning_rate * dm
        c -= learning_rate * dc

        # Compute and store cost
        cost = compute_cost(y, y_pred)
        costs.append(cost)

        # Print updated parameters
        print(f"Iteration {i+1}: m = {m}, c = {c}, Cost = {cost}")

    return m, c, costs

m, c, costs = gradient_descent(x, y)
```

```python


# Plot the cost function over iterations
plt.plot(range(1, len(costs) + 1), costs)
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.title('Cost Function Over Iterations')
plt.show()

# Plot the data points and the final regression line
plt.scatter(x, y, color='blue', label='Data Points')
plt.plot(x, m * x + c, color='red', label='Regression Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression with Gradient Descent')
plt.legend()
plt.show()

print(f"Final m: {m}, Final c: {c}")
```
